<h1>7. How do I filter rows of a pandas DataFrame by column value?</h1>

```python
movie = pd.read_csv("http://bit.ly/imdbratings")
movie.head()
```

```python
movie.shape
```

```python
type(False)
```

```python
#movies.duration is of type pandas Series

# method 1:

booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)

booleans[0:5]
```

```python
len(booleans)
```

```python
#convert booleans list into pandas series
is_long = pd.Series(booleans) 
```

```python
is_long.head()
```

```python
movies[is_long].head() #shows rows of dataframe with duration>=200
```

<h1> How do I use the "axis" parameter in pandas?</h1>

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
# drop the column called 'continent'
# axis = 1 means I want to drop a col

drinks.drop('continent',axis = 1).head()
```

```python
# drop 2nd row
# axis = 0 means I want to drop a row
drinks.drop(2,axis = 0).head()
```

```python
drinks.mean()
```

```python
# mean value of values across rows (or) from top to bottom
drinks.mean(axis=0) 
```

```python
# mean value of values across cols of each row (or) from left to right

# drinks.mean(axis=1) 
drinks.mean(axis='columns')
```

<h1> When should I use a "groupby" in pandas?</h1>

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
drinks.head()
```

```python
drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.beer_servings.mean()
```

```python
drinks[drinks.continent=='Europe'].beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.min()
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

```python
drinks.groupby('continent').mean()
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```
